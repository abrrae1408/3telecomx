```python
import pandas as pd
import numpy as np
import requests


url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
response = requests.get(url)
df = response.json()




```

```python
df_datos = pd.json_normalize(df, sep='_')
```

```python
# Vista previa de los datos
df_datos.tail()
```

##Transformación de los datos (ETL - T)
a. Eliminación de columnas irrelevantes:

```python
if 'customerID' in df_datos.columns:
    df_datos.drop('customerID', axis=1, inplace=True)
```

##Reemplazo de valores vacíos por NaN

```python
df_datos.replace(" ", np.nan, inplace=True)
```

##Conversión de tipos de datos (por ejemplo, account_changes_Total debería ser numérico)

```python
# Conversión de tipos de datos (por ejemplo, TotalCharges debería ser numérico)
# Se corrige el acceso a la columna 'account_Charges_Total' para que sea desde el DataFrame df_datos
df_datos['account_Charges_Total'] = pd.to_numeric(df_datos['account_Charges_Total'], errors='coerce')
```

##Relleno de valores nulos:

```python
df_datos['account_Charges_Total'].fillna(df_datos['account_Charges_Total'].median(), inplace=True)
```

##Conversión de variables categóricas:

```python
for col in df_datos.select_dtypes(include='object').columns:
    df_datos[col] = df_datos[col].astype('category')
```

##Carga de los datos limpios (ETL - L)

```python
df_datos.to_csv("TelecomX_Clean.csv", index=False)
```

#Visualización estratégica (solo con pandas y numpy)

```python
import matplotlib.pyplot as plt

# Distribución de Churn
df_datos['Churn'].value_counts().plot(kind='bar', title='Distribución de Cancelaciones')
plt.show()

# Relación entre TotalCharges y MonthlyCharges
# Se corrige el DataFrame utilizado para la visualización de df a df_datos
df_datos.plot(kind='scatter', x='account_Charges_Monthly', y='account_Charges_Total', alpha=0.5, c=df_datos['Churn'].cat.codes)
plt.title('MonthlyCharges vs TotalCharges (Coloreado por Churn)')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.show()
```

##Fase 1: Visualización Estratégica Avanzada con Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración estética
sns.set(style="whitegrid")

# 1. Distribución de cancelaciones
plt.figure(figsize=(6,4))
sns.countplot(data=df_datos, x="Churn", palette= ("tab10"))
plt.title("Distribución de Cancelaciones de Clientes")
plt.xlabel("¿Canceló el servicio?")
plt.ylabel("Cantidad de clientes")
plt.show()

```

```python
# 2. Cancelaciones por tipo de contrato
plt.figure(figsize=(8,5))


sns.countplot(data=df_datos, x="account_Contract", hue="Churn", palette="tab10")
plt.title("Cancelación por Tipo de Contrato")
plt.xlabel("Tipo de Contrato")
plt.ylabel("Clientes")
plt.xticks(rotation=15)
plt.show()

```

```python
 #3. Cargos mensuales según cancelación
plt.figure(figsize=(8,5))
sns.boxplot(data=df_datos, x="Churn", y="account_Charges_Monthly", palette="coolwarm") # Cambiado 'MonthlyCharges' a 'account_Charges_Monthly'
plt.title("Distribución de Cargos Mensuales por Churn")
plt.show()
```

##📊 Fase 2: Visualización Interactiva con Plotly

```python
import plotly.express as px

# 1. Relación entre TotalCharges, MonthlyCharges y Churn
# Corrected DataFrame from df to df_datos
fig = px.scatter(df_datos, x="account_Charges_Monthly", y="account_Charges_Total", color="Churn",
                 title="Relación entre MonthlyCharges y TotalCharges por Churn",
                 opacity=0.6, size_max=10)
fig.show()

# 2. Histograma interactivo de tenure
# Corrected DataFrame from df to df_datos
# Cambiado 'tenure' a 'customer_tenure' para que coincida con el nombre de la columna
fig = px.histogram(df_datos, x="customer_tenure", color="Churn", nbins=30,
                   title="Distribución del Tiempo de Permanencia (tenure) por Cancelación")
fig.show()
```

##1. Problema

Alta tasa de cancelación de clientes (~27%), que impacta la rentabilidad de la empresa.



##2. Metodología
Extracción vía API (JSON).

Transformación de datos: limpieza, conversión tipológica, tratamiento de nulos.

Carga en un DataFrame para análisis.

Visualización avanzada (Seaborn, Plotly).

Análisis de correlaciones y patrones por variables críticas.



##3. Principales Hallazgos
Los clientes con contratos mensuales son significativamente más propensos a cancelar.

Aquellos con cargos mensuales altos (> $80) muestran mayor probabilidad de churn.

Usuarios con bajo tiempo de permanencia (tenure < 12 meses) tienen alta tasa de cancelación.

Métodos de pago electrónicos se asocian a una mayor pérdida de clientes.

El género y la presencia de dependientes no son determinantes del churn.

##5. Recomendaciones de Negocio
